```python
#sector_symbol=["XLC","XLP","XLY","XLE","XLF","XLV", "XLI", "XLB", "XLRE", "XLK","XLU"]
#sector_name=["Communication Services","Consumer Staples", "Consumer Discretionary", "Energy", "Financial Sector", 
 #            "Health Care","Industrial Sector", "Material", "Real Estate", "Technology", "Utilities"]
#index_weight=[11.2, 6.0, 12.7, 2.7, 11.5, 12.8, 8.7, 2.7, 2.5, 26.7, 2.6]

data = {"Symbol":["XLC","XLP","XLY","XLE","XLF","XLV", "XLI", "XLB", "XLRE", "XLK","XLU"], "Sector": ["Communication Services","Consumer Staples", "Consumer Discretionary", "Energy", "Financial Sector", 
             "Health Care","Industrial Sector", "Material", "Real Estate", "Technology", "Utilities"],"index_weight": [11.2, 6.0, 12.7, 2.7, 11.5, 12.8, 8.7, 2.7, 2.5, 26.7, 2.6]}
df = pd.DataFrame(data)
df.sort_values(by=["index_weight"], ascending=False)
```

```python
labels = "Communication Services","Consumer Staples", "Consumer Discretionary", "Energy", "Financial Sector", "Health Care","Industrial Sector", "Material", "Real Estate", "Technology", "Utilities"
sizes = [11.2, 6.0, 12.7, 2.7, 11.5, 12.8, 8.7, 2.7, 2.5, 26.7, 2.6]
explode = (0.05, 0.05, 0.05, 0.05,0.05, 0.05, 0.05, 0.05, 0.05, 0.05, 0.05) 


fig1, ax1 = plt.subplots(figsize=(10, 8), subplot_kw=dict(aspect="equal"))

ax1.pie(sizes, explode=explode,  autopct='%1.1f%%', shadow=False, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

ax1.legend(labels,
          title="Sectors",
          loc="center left",
          bbox_to_anchor=(0.8, 0.5))


ax1.set_title("Sector Breakdown, Sector Weight")
plt.savefig('sectors.png')
plt.show()

```
